```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
team_basic_info = []

for i in range(1996, 2018):
    url = "https://www.sports-reference.com/cbb/seasons/" + str(i) + "-school-stats.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    stats_table = soup.find(id="all_basic_school_stats")
    teams = stats_table.find_all("tr")

    for team in teams:
        team = team.find_all("td")

        row = {}

        try:
            row['season'] = i
            row['school'] = team[0].get_text()
            row['total_games'] = team[1].get_text()#find('a')['href']
            row['wins'] = team[2].get_text()
            row['losses'] = team[3].get_text()
            row['w-l_perc'] = team[4].get_text()
            row['srs'] = team[5].get_text()
            row['sos'] = team[6].get_text()
            row['conf_w'] = team[7].get_text()
            row['conf_l'] = team[8].get_text()
            row['home_w'] = team[9].get_text()
            row['home_l'] = team[10].get_text()
            row['away_w'] = team[11].get_text()
            row['away_l'] = team[12].get_text()
            row['team_pts'] = team[13].get_text()
            row['opp_pts'] = team[14].get_text()
            row['separator'] = team[15].get_text()
            row['mp'] = team[16].get_text()
            row['fg'] = team[17].get_text()
            row['fga'] = team[18].get_text()
            row['fg_perc'] = team[19].get_text()
            row['3pt'] = team[20].get_text()
            row['3pta'] = team[21].get_text()
            row['3pt_perc'] = team[22].get_text()
            row['off_reb'] = team[23].get_text()
            row['tot_reb'] = team[24].get_text()
            row['assists'] = team[25].get_text()
            row['steals'] = team[26].get_text()
            row['blocks'] = team[27].get_text()
            row['turnovers'] = team[28].get_text()
            row['pers_fouls'] = team[29].get_text()
            
            team_basic_info.append(row)

        except:
            print("Ran into header row.")
```

```python
team_basic_df = pd.DataFrame(team_basic_info)
```

```python
cols = team_basic_df.columns.drop('school')
team_basic_df[cols] = team_basic_df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
team_basic_df = team_basic_df.drop('separator', axis=1)
```

```python
team_basic_df.to_csv('team_basic.csv')
```
